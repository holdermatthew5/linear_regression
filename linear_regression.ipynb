```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
insurance = pd.read_csv('./insurance.csv')
```

```python
insurance.head()
```

```python
insurance.describe()
```

```python
insurance.info()
```

```python
y = insurance['bmi'].values
x = insurance['age'].values
```

```python
x = x.reshape(-1, 1)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.8, test_size=.2, random_state=100)
```

```python
plt.scatter(x_train, y_train, color='yellow')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Mass by Age')
plt.show()
```

```python
lm = LinearRegression()
lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)
print(f'Train Accuracy {round(lm.score(x_train, y_train)* 100,2)}%')
print(f'Test Accuracy {round(lm.score(x_test, y_test)* 100,2)}%')
```

1. Getting older does not mean it's harder to maintain a proper BMI.
2. Being younger does not make it any easier to maintain a healthy weight.
3. Most peoples body mass index exists outside of what is considered healthy by the National Institutes of Health and this does not change with age.
